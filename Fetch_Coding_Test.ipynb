```python
import sqlite3
import pandas as pd
```

```python
brands = pd.read_csv('brands.csv')
receipt_items = pd.read_csv('receipt_items.csv')
receipts = pd.read_csv('receipts.csv')
users = pd.read_csv('users.csv')
```

```python
brands_conn = sqlite3.connect('brands.db')
receipt_items_conn = sqlite3.connect('receipt_items.db')
receipts_conn = sqlite3.connect('receipts.db')
users_conn = sqlite3.connect('users.db')
```

```python
brands.to_sql('brands', brands_conn, if_exists='append', index = False)
receipt_items.to_sql('receipt_items', receipt_items_conn, if_exists='append', index = False)
receipts.to_sql('receipts', receipts_conn, if_exists='append', index = False)
users.to_sql('users', users_conn, if_exists='append', index = False)
```
